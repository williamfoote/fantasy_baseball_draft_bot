# Now on to the fun stuff: modeling

```python

labels = joined['overallPickNumber']
# Remove the labels from the features
# axis 1 refers to the columns
#features= joined.drop(['overallPickNumber', 'Unnamed: 0', 'ESPNNAME', 'ESPNID', 'fullName', 'team name', 'player_id'], axis = 1)
features = joined[['pre_draft_ranking', 'adp', 'team_id', 'player_id', 'pr', '1B', '2B','3B', 'SS', 'C', 'LF', 'CF', 'RF', 'SP', 'RP', 'HR', 'RBI', 'RC', 'R', 'IP', 'QS', 'SVHD', 'count_slots']]
# Saving feature names for later use
feature_list = ['pre_draft_ranking', 'adp', 'team_id', 'player_id', 'pr', '1B', '2B','3B', 'SS', 'C', 'LF', 'CF', 'RF', 'SP', 'RP', 'HR', 'RBI', 'RC', 'R', 'IP', 'QS', 'SVHD', 'count_slots']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import HistGradientBoostingRegressor
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
train_features
```

```python
# Instantiate model with 1000 decision trees
hbg = HistGradientBoostingRegressor()
# Train the model on training data
est = hbg.fit(train_features, train_labels)
```

```python
# Use the forest's predict method on the test data
predictions = hbg.predict(test_features)
# Calculate the absolute errors
errors = abs(predictions - test_labels)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'picks.')
```

```python
from sklearn.model_selection import cross_val_score
from sklearn import metrics
def modelfit(alg, predictor_df, predictors, label_df, label_var, performCV=True, printFeatureImportance=True, cv_folds=5):
    #Fit the algorithm on the data
    alg.fit(predictor_df[predictors], label_df)
        
    #Predict training set:
    dtrain_predictions = alg.predict(predictor_df[predictors])
    # dtrain_predprob = alg.predict_proba(predictor_df[predictors])[:,1]
    
    #Perform cross-validation:
    if performCV:
        cv_score = cross_val_score(alg, predictor_df[predictors], label_df, cv=cv_folds, scoring='neg_root_mean_squared_error')
    
    #Print model report:
    print ("\nModel Report")
    print ("MSE : %.4g" % metrics.mean_squared_error(label_df.values, dtrain_predictions))
    #print ("AUC Score (Train): %f" % metrics.roc_auc_score(predictor_df[label_var], dtrain_predprob))
    
    if performCV:
        print ("CV Score : Mean - %.7g | Std - %.7g | Min - %.7g | Max - %.7g" % (np.mean(cv_score),np.std(cv_score),np.min(cv_score),np.max(cv_score)))
        
    #Print Feature Importance:
    if printFeatureImportance:
        feat_imp = pd.Series(alg.feature_importances_, predictors).sort_values(ascending=False)
        feat_imp.plot(kind='bar', title='Feature Importances')
        plt.ylabel('Feature Importance Score')
        
```

```python
from sklearn.model_selection import GridSearchCV
#Choose all predictors except target & IDcols
predictors = joined[['pre_draft_ranking', 'team_id', 'defaultPositionId', 'adp', 'R', 'HR', 'OPS', 'QS', 'SVHD']]
labels = joined['overallPickNumber']
train_features, test_features, train_labels, test_labels = train_test_split(predictors, labels, test_size = 0.25, random_state = 42)
hbg = HistGradientBoostingRegressor()
param_grid = {
    'learning_rate': (0.01, 0.1, 1, 10),
    'max_leaf_nodes': (3, 10, 30)}
model_grid_search = GridSearchCV(hbg, param_grid=param_grid,
                                 n_jobs=2, cv=5)
model_grid_search.fit(train_features, train_labels)
```

```python
accuracy = model_grid_search.score(test_features, test_labels)
print(
    f"The test accuracy score of the grid-searched pipeline is: "
    f"{accuracy:.2f}"
)
```

```python
cv_results = pd.DataFrame(model_grid_search.cv_results_).sort_values(
    "mean_test_score", ascending=False)
cv_results.head()
```
